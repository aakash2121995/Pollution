```python
def getAbbr(station):
    list1 = []
    for a in station:
        pos1 = a.find('(')
        pos2 = a.find(')')
        list1.append(a[pos1: pos2+1])
    
    return list1
```

```python
civmalist = getAbbr(civma.columns[1:])
#display(civmalist)
itolist = getAbbr(ito.columns[1:])
#display(itolist)
civmalist.extend(itolist)
listcols = list(set(civmalist))
display(listcols, len(listcols))
```

```python
def checkDate(set_, dateTime):
   
    if dateTime in set_:
        return True
    else:
        return False
```

```python
def checkCol(col, allCols):
    for a in allCols:
        if a.find(col) >=0:
            return True, a
    return False, -1
```

```python
emptyDic = {}
for ele in listcols:
    emptyDic[ele] = []
sinDic = {}
for ele in listcols:
    sinDic[ele] = [np.nan]
```

```python
emptyDic["Time"] = []
sinDic["Time"] = [np.nan]
```

```python
# in_c = {}
in_cm = {}
in_i = {}

# set_c = set(civil.index)
set_cm = set(civma.index)
set_i = set(ito.index)

for col in listcols:
    status1, colName1 = checkCol(col, civma.columns)
    in_cm[col] = [status1, colName1]
#     status2, colName2 = checkCol(col, mandircols)
#     in_m[col] = [status2, colName2] 
    status3, colName3 = checkCol(col , ito.columns)
    in_i[col] = [status3, colName3]
```

```python
start = datetime(2011,1,1,0,0,0)
combinedDf = pd.DataFrame(emptyDic)
i = 0

while start < datetime.today():
    
    currDate = start.strftime("%Y-%m-%d %H:%M:%S")
    # checking if date exists or not,,
    tempDf = pd.DataFrame(sinDic)
#     display(tempDf)
    cm_status = checkDate(set_cm, start)
    i_status = checkDate(set_i, start)
    
    if i%1000 == 0:
        display(tempDf)
        display(combinedDf.tail())
        display(currDate)
    
    if cm_status or i_status:
        tempDf.loc[0,"Time"] = currDate
        for col in listcols:
#             print(col)
    
            status1, colName1 = in_cm[col][0], in_cm[col][1]
            status2, colName2 = in_i[col][0], in_i[col][1]
            
            if cm_status and status1 and (pd.isnull(civma.ix[currDate][colName1]) or 
                                         (civma.ix[currDate][colName1] < 0 and col != "(VWS)"))  :
                status1 = False
            if i_status and status2 and (pd.isnull(ito[currDate][colName2][0]) or 
                                        (ito[currDate][colName2][0] < 0 and col != "(VWS)")):
                status2 = False
            
#             print(colName1)
#             print(colName2)
            if status1 and status2 and cm_status and i_status :
            
                tempDf.loc[0,col] = (civma.ix[currDate][colName1] + ito[currDate][colName2][0])/2
            elif status1 and cm_status:
            
                tempDf.loc[0,col] = civma.ix[currDate][colName1]
            
            elif status2 and i_status:
                tempDf.loc[0,col] = ito[currDate][colName2][0]
        
            
#             combinedDf = combinedDf.append(tempDf,)
#     print("appending")
        combinedDf = pd.concat([combinedDf,tempDf],ignore_index=True)
    
    
    i = i +1
       
    start = start + timedelta(hours=1)
```

```python
combinedDf.set_index("Time",inplace=True)
```
