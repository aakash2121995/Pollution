```python
display(civil.columns)
civil = civil.rename(columns = {'Civil Lines/PM10(RPM)' : 'Civil Lines/PM10(PM10)' })
civil = civil.rename(columns = {'Civil Lines/WindDirection(Wind Dir S)' : 'Civil Lines/WindDirection(WD)' })
civil = civil.rename(columns = {'Civil Lines/Toluene(Toluene)' : 'Civil Lines/Toluene(TOL)' })
civil = civil.rename(columns = {'Civil Lines/OZONE(OZONE)' : 'Civil Lines/OZONE(O3)' })
civil = civil.rename(columns = {'Civil Lines/Benzene(Benzene)' : 'Civil Lines/Benzene(BEN)' })
civil = civil.rename(columns = {'Civil Lines/Bar Pressure(Bar Pressure)' : 'Civil Lines/Bar Pressure(BP)' })
civil = civil.rename(columns = {'Civil Lines/PM2.5(Dust (PM2.5) )' : 'Civil Lines/PM2.5(PM2.5)' })
display(civil.columns)
```

```python
display(mandir.columns)
mandir = mandir.rename(columns = {'Mandir Marg/Vertical Wind Speed(Wind Speed V)' : 'Mandir Marg/Vertical Wind Speed(VWS)' })
mandir = mandir.rename(columns = {'Mandir Marg/Ozone(Ozone)' : 'Mandir Marg/Ozone(O3)'})
mandir = mandir.rename(columns = {'Mandir Marg/M & P Xylene(M_P_Xylene)' : 'Mandir Marg/M & P Xylene(MP-Xylene)'})
display(mandir.columns)
```

```python
display(ito.columns)
ito = ito.rename(columns = {'ITO/Ozone(OZONE)' : 'ITO/Ozone(O3)'})
```

```python
def getAbbr(station):
    list1 = []
    for a in station:
        pos1 = a.find('(')
        pos2 = a.find(')')
        list1.append(a[pos1: pos2+1])
    
    return list1
```

```python
civilcols = civil.columns[1:]
mandircols = mandir.columns[1:]
itocols = ito.columns[1:]
display(civilcols)
display(mandircols)
display(itocols)
```

```python
civillist = getAbbr(civilcols)
mandirlist = getAbbr(mandircols)
itolist = getAbbr(itocols)
#display(punjabilist)
#display(shadilist)
civillist.extend(mandirlist)
civillist.extend(itolist)
listCols = list(set(civillist))
display(listCols)
display(len(listCols))
```

```python
def checkDate(set_, dateTime):
   
    if dateTime in set_:
        return True
    else:
        return False
```

```python
def checkCol(col, allCols):
    for a in allCols:
        if a.find(col) >=0:
            return True, a
    return False, -1
```

```python
emptyDic = {}
for ele in listCols:
    emptyDic[ele] = []
sinDic = {}
for ele in listCols:
    sinDic[ele] = [np.nan]
```

```python
emptyDic["Time"] = []
sinDic["Time"] = [np.nan]
```

```python
in_c = {}
in_m = {}
in_i = {}

set_c = set(civil.index)
set_m = set(mandir.index)
set_i = set(ito.index)

for col in listCols:
    status1, colName1 = checkCol(col, civilcols)
    in_c[col] = [status1, colName1]
    status2, colName2 = checkCol(col, mandircols)
    in_m[col] = [status2, colName2] 
    status3, colName3 = checkCol(col , itocols)
    in_i[col] = [status3, colName3]
```

```python
start = datetime(2011,1,1,0,0,0)
combinedDf = pd.DataFrame(emptyDic)
i = 0

while start < datetime.today():
    
    currDate = start.strftime("%Y-%m-%d %H:%M:%S")
    # checking if date exists or not,,
    tempDf = pd.DataFrame(sinDic)
#     display(tempDf)
    c_status = checkDate(set_c, start)
    m_status = checkDate(set_m, start)
    
    if i%1000 == 0:
        display(tempDf)
        display(combinedDf.tail())
        display(currDate)
    
    if c_status or m_status:
        tempDf.loc[0,"Time"] = currDate
        for col in listCols:
#             print(col)
    
            status1, colName1 = in_c[col][0], in_c[col][1]
            status2, colName2 = in_m[col][0], in_m[col][1]
            
            if c_status and status1  and (pd.isnull(civil[currDate][colName1][0]) or 
                                         (civil[currDate][colName1][0] < 0 and col != "(VWS)"))  :
                status1 = False
            if m_status and status2 and (pd.isnull(mandir[currDate][colName2][0]) or 
                                         (mandir[currDate][colName2][0] < 0 and col != "(VWS)")):
                status2 = False
            
#             print(colName1)
#             print(colName2)
            if status1 and status2 and c_status and m_status :
            
                tempDf.loc[0,col] = (civil[currDate][colName1][0] + mandir[currDate][colName2][0])/2
            elif status1 and c_status:
            
                tempDf.loc[0,col] = civil[currDate][colName1][0]
            
            elif status2 and m_status:
                tempDf.loc[0,col] = mandir[currDate][colName2][0]
        
            
#             combinedDf = combinedDf.append(tempDf,)
#     print("appending")
        combinedDf = pd.concat([combinedDf,tempDf],ignore_index=True)
    
    
    i = i +1
       
    start = start + timedelta(hours=1)
```

```python
combinedDf.set_index("Time",inplace=True)
```
