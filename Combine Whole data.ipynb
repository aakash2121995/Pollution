```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
from aakashLib import *
import glob
%matplotlib inline
```

```python
files = glob.glob("TrueData/*sv")

firstCsv = pd.read_csv(files[2],sep='\t',parse_dates=["Time"])
# files

files.remove('TrueData/IGI Airport.tsv')
files.remove('TrueData/CivilMandir 27April.csv')
files

# firstCsv = pd.read_csv(files[0])    
for file in files:
    print("Joining " + file)
    nextCsv = pd.read_csv(file,parse_dates=["Time"])
    station = file.split('/')[1].split('.')[0]
    display(nextCsv.shape)
    plotMissingVals(nextCsv)
    
    for col in nextCsv.columns[1:]:
        nextCsv.rename(columns={col:station + ' ' + col},inplace=True)
#     display(nextCsv.columns)
    firstCsv = firstCsv.merge(nextCsv,how='outer')
    
display(firstCsv.shape)
plotMissingVals(firstCsv)
```
