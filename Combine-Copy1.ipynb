```python
import pandas as pd
import os
import glob
```

```python
def mergeAll(csvList, folder):
    stname = folder[:-1]
#     print "station name is", stname
    df1 = pd.read_csv(csvList[0])
    df1.rename(columns={'Values': csvList[0][:-4]}, inplace=True)
    df1 = df1.fillna('-1')
    df1[df1.columns[2]] = df1[df1.columns[2]].astype(str)
    df1['Time'] = df1["Date"].map(str) + ' ' + df1["Time"]
    df1 = df1.drop('Date',axis=1)
    df1 = df1.drop_duplicates(subset="Time")
    
#     print df1["Time"].value_counts()
    
#     print(df1.head)
#     print df1.shape
    for csv in csvList[1:]:
        print "Merging " + csv
        df2 = pd.read_csv(csv)
        if df2.shape[0] == 0:
            continue
        df2.rename(columns={'Values': csv[:-4]}, inplace=True)
        df2 = df2.fillna('-1')
        df2[df2.columns[2]] = df2[df2.columns[2]].astype(str)
        df2['Time'] = df2["Date"].map(str) + ' ' + df2["Time"]
        df2 = df2.drop('Date',axis=1)
        df2 = df2.drop_duplicates(subset="Time")
        print("df2 shape{0}".format(df2.shape))
        print df2.as_matrix()[0]
        valDf1 = df1[df1.columns[0]].as_matrix().tolist()
        valDf2 = df2[df2.columns[0]].as_matrix().tolist()
        
        valDf1.extend(valDf2)
        
        valDf1 = set(valDf1)
        valDf1 = list(valDf1)
        
        
        print len(valDf1)
#         print  valDf1[0]
        
        df1 = df1.merge(df2,how='outer')
        print df1.shape
    return df1
#     df1.to_csv(stName + ".csv")

```

```python
print "Reading folders ..."
df = pd.DataFrame()

# df.m
# df.merge()
folders  = glob.glob("*/")
print folders
print
for f in folders:
    csvName = glob.glob(f+"*.csv")
    print "Creating " + f
    df = mergeAll(csvName, f)
    outPutFolder = "Combined Data/"
    print "Saving " + f
    df.to_csv(outPutFolder +  f[:-1]+'.csv',index=False)
```
