```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("AvgStep3.csv")
```

```python
def getDataLOL(data):
    print(data.shape)
    temp = data[data.year>=2015]
    data = data[data.year<2015]
    trainData = data[data.year<2012]
    test = data[data.year>=2012]
    valData = test[test.year<2014]
    testData = test[test.year>=2014]
    trainData = trainData.append(temp)
    return trainData,valData,testData
def getData(data):
#     print(data.shape)
#     temp = data[data.year>=2015
    trainData = data[data.year<2011]
    test = data[data.year>=2011]
    valData = test[test.year<2013]
    testData = test[test.year>=2013]
#     trainData = trainData.append(temp)
    return trainData,valData,testData
```

```python
#drop waste features
data = data.drop(['ndDate','WBAN ','STN---','  GUST'],axis = 1)
```

```python

def Cel(y):
    return (y-32)*5/9

data['ndTemp'] = Cel(data['ndTemp'])
for col in data.columns:
    if col[:6] == '   MAX' or col[:5] == '  MIN' or col[:8] == '   TEMP':
        data[col] = Cel(data[col])
```

```python
from sklearn.preprocessing import LabelEncoder as LE
for col in data.columns:
    if col[:5] == 'prcpt':
        print(col)
        data[col][data[col] == -1] = 'N'
        data[col] = LE().fit_transform(data[col])
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

trainData,valData,testData = getData(data)
X_train = trainData.drop('ndTemp',axis=1)
y_train = trainData.ndTemp

X_val = valData.drop('ndTemp',axis=1)
y_val = valData.ndTemp

X_test = testData.drop('ndTemp',axis=1)
y_test = testData.ndTemp


from xgboost import XGBRegressor 
for i in range(300,1200,10):
    print(i)
    xgb = XGBRegressor(n_estimators=i)
    xgb.fit(X_train,y_train)

    xgbTrain = xgb.predict(X_train)
    print(mae(xgbTrain,y_train))
    print(np.sqrt(mse(xgbTrain,y_train)))

    xgbTest = xgb.predict(X_test)
    print(mae(xgbTest,y_test))
    print(np.sqrt(mse(xgbTest,y_test)))

    xgbVal = xgb.predict(X_val)
    valRmse = np.sqrt(mse(xgbVal,y_val))
    print(mae(xgbVal,y_val))
    print(valRmse)
```
